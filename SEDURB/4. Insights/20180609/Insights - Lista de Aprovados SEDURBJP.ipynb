### 1. Importando Bibliotecas

```python
# OBS: Algumas bibliotecas e comandos podem eventualmente não serem utilizados

# Bibliotecas de analise
import pandas as pd
import numpy as np

# Bibliotecas de visualozação 
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline
```

### 2. Localizando e abrindo arquivo

```python
df_sedurb = pd.read_csv('Lista parcial de Aprovados na SEDURB.csv')
```

### 3. Inspecionando o arquivo

```python
df_sedurb.head() # Pode se ver que existe uma coluna desnecessária, vamos elimina-la
```

### 4. Tratando dos dados

```python
# Eliminando a coluna desnecessária 

df_sedurb.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
df_sedurb.head()
```

### 5. Insights

```python
df_sedurb['Notas'].max() # A maior nota tirada no certame da SEDURB
```

```python
df_sedurb['Notas'].min() # A menor nota tirada no certame da SEDURB
```

```python
df_sedurb['Notas'].describe() # Resumo dos dados
```

```python
df_sedurb['Notas'].unique() # Notas todas as variações de notas tiradas no certame
```

```python
# Podemos verificar com este comando que somente os candidatos que obtiveram nota superior ou igual a 61,50 
# se classificaram para a próxima fase do concurso
df_sedurb.sort_values('Notas', ascending=False).head(266)
                                                         
```

### 6. Salvando os insights

```python
df_candidatos_aptos = df_sedurb.sort_values('Notas', ascending=False).head(266)
```
